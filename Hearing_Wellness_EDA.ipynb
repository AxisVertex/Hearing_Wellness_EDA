```python

# Age group: Histogram
# Age_Group and Hearing_FOMO: Bar
# Age_Group and Daily_Headphone_Use: Bar
# Age_Group and Interest_in_Hearing_App: bar
# Age_Group and Belief_Early_Hearing_Care: :bar

# Ear_Discomfort and Daily_Headphone_Use: Bar

# Interest_in_Hearing_App: Pie chart
# Paid_App_Test_Interest: Pie chart
# Awareness_on_hearing_and_Willingness_to_invest: Pie chart
```

```python
# Distribution of age group
age_order = ["Under 18", "18 - 24", "25 - 34", "35 - 44", "45 - 54", "55 - 64"]
sns.countplot(x="Age_group", data=df, order=age_order)
plt.title("Distribution of Age Group")
plt.xlabel("Age Group")
plt.ylabel("Count")
plt.show()
```

```python

df.groupby(["Age_group", "Hearing_FOMO"]).size().unstack(fill_value=0).plot(kind="bar")
plt.xlabel("Age Group")
plt.ylabel("Number of respondents")
plt.title("Age Group vs. Hearing FOMO")
plt.show()
# age_fomo = df.groupby("Age_group")["Hearing_FOMO"].sum().plot(kind="bar")
```

```python
df.groupby(["Daily_Headphone_Use", "Age_group"]).size().unstack(fill_value=0).plot(kind="bar", stacked=True)
plt.title("Daily Headphone Use vs. Age Group")
plt.xlabel("Hours of Use")
plt.ylabel("Number of respondents")
plt.show()
```

```python
df.groupby(["Age_group", "Interest_in_Hearing_App"]).size().unstack(fill_value=0).plot(kind="bar")
plt.title("Interest in Hearing App Across Age Groups")
plt.xlabel("Age Group")
plt.ylabel("Number of respondents")
plt.show()
```

```python
df.groupby(["Age_group", "Belief_Early_Hearing_Care"]).size().unstack(fill_value=0).plot(kind="bar")
plt.title("Belief in Early Hearing Care Between Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of respondents")
plt.legend(["Strongly disagree", "Disagree", "Neutral", "Agree", "Strongly Agree"])
plt.show()
```

```python
ordering = ["Less than 1 hour", "1-2 hours", "2-4 hours", "More than 4 hours"]
df["Daily_Headphone_Use"] = pd.Categorical(df["Daily_Headphone_Use"], categories=ordering, ordered=True)

df.groupby(["Daily_Headphone_Use", "Ear_Discomfort_After_Use"]).size().unstack(fill_value=0).plot(kind="bar")
plt.title("Daily Use of Headphones vs. Ear Discomfort after Use")
plt.xlabel("Hours of use")
plt.ylabel("Number of respondents")
plt.show()
```

```python

# Interest_in_Hearing_App: Pie chart
# Paid_App_Test_Interest: Pie chart
# Awareness_on_hearing_and_Willingness_to_invest: Pie chart
```

```python
interested_in_app = df["Interest_in_Hearing_App"].value_counts()
# interested_in_app
plt.pie(interested_in_app, labels=interested_in_app.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of the People who are interested in a Hearing App")
plt.show()
```

```python
interested_in_paid = df["Paid_App_Test_Interest"].value_counts()
# interested_in_app
plt.pie(interested_in_paid, labels=interested_in_paid.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of the People who are interested in a Paid Hearing App")
plt.show()
```

```python
investing = df["Awareness_on_hearing_and_Willingness_to_invest"].value_counts()
# interested_in_app
plt.pie(investing, labels=investing.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of the People who are Interested in Investing In Hearing App")
plt.show()
```

```python
hearing_test_barrier = df["Hearing_Test_Barrier"].value_counts().head().plot(kind="bar")
plt.title("Top 5 Hearing Test Barriers")
plt.xlabel("Barrier")
plt.ylabel("Count")
plt.show()
```
